```python
# This function is to checks if there is at least a comma within the text
# Then add double qoutes at the beginning and the end of the text if not exists
# To be used as text qualifier
def quote_if_needed(text):
    if not isinstance(text, str):
        return text  # Handle non-string values gracefully

    # Remove internal quotes
    text = text.replace('"', '')

    # If the text contains a comma, we may need to wrap it in quotes
    if ',' in text:
        if not text.startswith('"'):
            text = '"' + text
        if not text.endswith('"'):
            text = text + '"'
    return text
```

```python
# This Cell Applies the function for all the features with object dtype 
# Select the object columns to apply
object_columns = BR.select_dtypes(include=['object']).columns

# apply the function to the selected columns
for col in object_columns:
    BR[col] = BR[col].apply(quote_if_needed)
    CA[col] = CA[col].apply(quote_if_needed)
    US[col] = US[col].apply(quote_if_needed)
```

```python
# Saving a Fixed Version for each file to be used.
BR.to_csv('BR_Fixed.csv', index=False)
CA.to_csv('CA_Fixed.csv', index=False)
US.to_csv('US_Fixed.csv', index=False)
```
